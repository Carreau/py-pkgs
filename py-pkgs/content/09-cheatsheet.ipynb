## Setup

1. Install [Cookiecutter](https://cookiecutter.readthedocs.io/en/1.7.2/):

```bash
$ pip install cookiecutter
```
2. Navigate to desired local directory and use the [UBC-MDS-Cookiecutter](https://github.com/UBC-MDS/cookiecutter-ubc-mds) to create a new Python package template:

```bash
$ cookiecutter https://github.com/UBC-MDS/cookiecutter-ubc-mds.git
```
3. Navigate to the newly created package directory and initialize a [poetry](https://python-poetry.org/) project:

```bash
$ cd mypkg
$ poetry init
```
4. Create a [new GitHub repository](https://github.com/new) (without a README).
5. Sync your local and remote repositories:

```bash
$ git init
$ git remote add origin git@github.com:USER/REPO_NAME.git
$ git add .
$ git commit -m "first commit"
$ git push -u origin master
```
6. Make changes to your package as desired.

## Testing

1. Install `pytest` and `pytest-cov` as development dependencies:

```bash
$ poetry add --dev pytest pytest-cov
```
2. Write tests using `pytest` syntax in `mypkg/tests/test_pypkgs.py` (see Section {ref}`04-pytest-tests`).
3. Run tests with:

```bash
$ poetry run pytest
```
4. Check test coverage with:

```bash
$ poetry run pytest --cov=mypkg tests/
```

## Documentation

1. Write NumPy-style docstrings where appropriate (see Section {ref}`05-code-docstring`).
2. Add `sphinx` and `sphinxcontrib-napoleon` as development dependencies:

```bash
$ poetry add --dev sphinx sphinxcontrib-napoleon
```
3. Add `extensions = ['sphinx.ext.napoleon', 'sphinx.ext.autodoc']` in the `conf.py` file in the `docs` directory if not already there.
4. Render documentation:

```bash
$ poetry run sphinx-apidoc -f -o docs/source mypkg
$ cd docs
$ poetry run make html
$ cd ..
```
5. Use [Read the Docs](https://readthedocs.org/) to host your documentation online (see Section {ref}`05-rendering-remotely`).
